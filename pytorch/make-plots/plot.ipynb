```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.unicode'] = True
matplotlib.rcParams['text.latex.preamble'] = [r'\usepackage{amsmath}']
matplotlib.rcParams['lines.linewidth'] = 1.5
matplotlib.rcParams.update({'font.size': 16, 
                            'axes.titlesize': 'large',
                            'axes.labelsize': 'large',
                            'legend.fontsize': 'large'})
```

## Figure 1: learning rate tuning

```python
fig, ax = plt.subplots(1,3, figsize=(12,3.5))
plt.setp(ax, xticks=[0,1,2,3,4], yticks=[0.0,0.5,1.0], xticklabels=['$10^0$','$10^{-1}$','$10^{-2}$','$10^{-3}$','$10^{-4}$'])

def metric_normalize(values):
    return (min(values) / values) * 1.0
    
# Madam
cifar_10_results = [20.24, 79.98, 90.46, 83.51, 71.89]
cifar_100_results = [10, 67.22, 69.88, 47.72, 19.57]
imagenet_results = [0.1, 4.5, 56.83, 54.59, 19.69]
gan_results = [255.7,115.8,29.16,39.28,83.74]
transformer_results = [3791,367.6,214.3,267.6,552.2]
cifar_10_results = 100 - np.array(cifar_10_results)
cifar_100_results = 100 - np.array(cifar_100_results) 
imagenet_results = 100 - np.array(imagenet_results)
gan_results = np.array(gan_results)
transformer_results = np.array(transformer_results)
cifar_10_results = metric_normalize(cifar_10_results)
cifar_100_results = metric_normalize(cifar_100_results)
imagenet_results = metric_normalize(imagenet_results)
gan_results = metric_normalize(gan_results)
transformer_results = metric_normalize(transformer_results)
x_axis = np.arange(len(cifar_10_results))
ax[0].plot(x_axis, cifar_10_results, label='CIFAR-10')
ax[0].plot(x_axis, cifar_100_results, label='CIFAR-100')
ax[0].plot(x_axis, imagenet_results, label='ImageNet')
ax[0].plot(x_axis, gan_results, label='CIFAR-10 cGAN')
ax[0].plot(x_axis, transformer_results, label='Wikitext-2')
ax[0].scatter([2,2,2],[1,1,1],marker='x',c='r',s=100,zorder=10,linewidth=3)
ax[0].set_title('Madam')

# SGD
cifar_10_results = [87.82, 92.58, 91.64, 86.79, 78.38]
cifar_100_results = [59.58, 70.01, 67.22, 58.81, 36.12]
imagenet_results = [39.95, 60.88, 47.1, 29.70, 3.8]
gan_results = [678.7,380.2,41.8,54.95,109.8]
transformer_results = [204.4,374.9,792.3,2005,16328]
cifar_10_results = 100 - np.array(cifar_10_results)
cifar_100_results = 100 - np.array(cifar_100_results) 
imagenet_results = 100 - np.array(imagenet_results)
gan_results = np.array(gan_results)
transformer_results = np.array(transformer_results)
cifar_10_results = metric_normalize(cifar_10_results)
cifar_100_results = metric_normalize(cifar_100_results)
imagenet_results = metric_normalize(imagenet_results)
gan_results = metric_normalize(gan_results)
transformer_results = metric_normalize(transformer_results)
x_axis = np.arange(len(cifar_10_results))
ax[1].plot(x_axis, cifar_10_results, label='CIFAR-10')
ax[1].plot(x_axis, cifar_100_results, label='CIFAR-100')
ax[1].plot(x_axis, imagenet_results, label='ImageNet')
ax[1].plot(x_axis, gan_results, label='CIFAR-10 cGAN')
ax[1].plot(x_axis, transformer_results, label='Wikitext-2')
ax[1].scatter([0,1,2],[1,1,1],marker='x',c='r',s=100,zorder=10,linewidth=3)
ax[1].set_title('SGD')

# Adam
cifar_10_results = [10, 91.36, 92.19, 93.28, 89.73]
cifar_100_results = [1, 61.18, 61.2, 70.47, 62.84]
imagenet_results = [0.1, 49.9, 67.4, 67.6, 63.2]
gan_results = [496.1,496.2,300,106.8,26.79]
transformer_results = [100000,25519,1526,258,199.7] #first value overflow
cifar_10_results = 100 - np.array(cifar_10_results)
cifar_100_results = 100 - np.array(cifar_100_results) 
imagenet_results = 100 - np.array(imagenet_results)
gan_results = np.array(gan_results)
transformer_results = np.array(transformer_results)
cifar_10_results = metric_normalize(cifar_10_results)
cifar_100_results = metric_normalize(cifar_100_results)
imagenet_results = metric_normalize(imagenet_results)
gan_results = metric_normalize(gan_results)
transformer_results = metric_normalize(transformer_results)
x_axis = np.arange(len(cifar_10_results))
ax[2].plot(x_axis, cifar_10_results, label='CIFAR-10')
ax[2].plot(x_axis, cifar_100_results, label='CIFAR-100')
ax[2].plot(x_axis, imagenet_results, label='ImageNet')
ax[2].plot(x_axis, gan_results, label='CIFAR-10 cGAN')
ax[2].plot(x_axis, transformer_results, label='Wikitext-2')
ax[2].scatter([3,4,4],[1,1,1],marker='x',c='red',s=100,zorder=10,linewidth=3)
ax[2].set_title('Adam')

for ax_idx in [0,1,2]:
    ax[ax_idx].set_xlabel('Learning rate $\eta$')
    ax[ax_idx].set_ylim(-0.05,1.05)
    ax[ax_idx].set_yticklabels([])
ax[0].set_yticklabels([0.0,0.5,1.0])
    
ax[2].legend(loc='upper center', bbox_to_anchor=(1.6, 1.0))
ax[0].set_ylabel('Best error $/$ error')

fig.tight_layout(w_pad=0.5)
plt.show()
fig.savefig('figures/lr-compare.pdf', bbox_inches='tight')
```
